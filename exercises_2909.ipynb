Input voltages:

$X=\{0,5\}$

State (membrane potential):

$S=\{-75,-70,-65\}$

Output voltages:

$Y=\{0,5\}$

Transition function:

$\delta(x,s):X\times S\to S$

$\delta(0,-75)=-75$

$\delta(0,-70)=-70$

$\delta(0,-65)=-65$

$\delta(5,-75)=-70$

$\delta(5,-70)=-65$

$\delta(5,-65)=-75$

Output function:

$\lambda:S\to Y$

$\lambda(-75)=0$

$\lambda(-70)=0$

$\lambda(-65)=5$

Time base (if the base unit is seconds and we want time to pass in milliseconds):

$c=0.001$

$T=cn$

$n\in\mathbb{N}$
